```python
#EDA


#Top 7 RICE PRODUCTION State Data(Bar_plot)

top_rice = (
    df.groupby('State Name')['RICE PRODUCTION (1000 tons)']
    .sum()
    .nlargest(7)
    .reset_index()
)

top_rice
```

```python
# Visualization of Top 5 RICE PRODUCTION States

plt.figure(figsize=(12, 6))

sns.barplot(
    x='RICE PRODUCTION (1000 tons)', 
    y='State Name', 
    data=top_rice, 
    palette='viridis'
)

plt.title('Top 7 Rice Producing States', fontsize=18, weight='bold')
plt.xlabel('Rice Production (1000 tons)', fontsize=14, weight='bold')
plt.ylabel('State Name', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')

plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

top_wheat = (
    df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)']
    .sum()
    .nlargest(5)
    .reset_index()
)

top_wheat
```

```python

# Bar chart for top 5 wheat-producing states
plt.figure(figsize=(12, 6))

sns.barplot(
    x='WHEAT PRODUCTION (1000 tons)', 
    y='State Name', 
    data=top_wheat, 
    palette='coolwarm'
)

plt.title('Top 5 Wheat Producing States', fontsize=18, weight='bold')
plt.xlabel('Wheat Production (1000 tons)', fontsize=14, weight='bold')
plt.ylabel('State Name', fontsize=14, weight='bold')
plt.show()

# Calculate percentage share of wheat production
total_wheat_production = top_wheat['WHEAT PRODUCTION (1000 tons)'].sum()
top_wheat['Percentage'] = (top_wheat['WHEAT PRODUCTION (1000 tons)'] / total_wheat_production) * 100

# Pie chart to visualize percentage share
plt.figure(figsize=(6, 6))

plt.pie(
    top_wheat['Percentage'], 
    labels=top_wheat['State Name'], 
    explode=[0.01] * len(top_wheat), 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette('coolwarm', n_colors=5)
)

plt.title('Percentage Share of Wheat Production by Top 5 States', fontsize=18, weight='bold')
plt.show()
```

```python
# Soybean Production by Top 5 States and Yield Efficiency
soyabean = df.loc[:, ['State Name', 'SOYABEAN PRODUCTION (1000 tons)', 'SOYABEAN AREA (1000 ha)']]

soyabean
```

```python
# Calculate Soybean Yield Efficiency
soyabean['Yield Efficiency (tons/hectare)'] = (
    soyabean['SOYABEAN PRODUCTION (1000 tons)'] / soyabean['SOYABEAN AREA (1000 ha)']
)

soyabean[['State Name', 'Yield Efficiency (tons/hectare)']]
```

```python
# Grouped Soybean Production Data by State
soyabean_grouped = (
    soyabean.groupby('State Name')
    .agg({
        'SOYABEAN PRODUCTION (1000 tons)': 'sum',
        'SOYABEAN AREA (1000 ha)': 'sum',
        'Yield Efficiency (tons/hectare)': 'mean'
    })
    .reset_index()
)

soyabean_grouped
```

```python
# Top 5 States for Soybean Production
top_5_soyabean = (
    soyabean_grouped.sort_values(by='SOYABEAN PRODUCTION (1000 tons)', ascending=False)
    .head(5)
)

top_5_soyabean
```

```python
# Visualization: Top 5 States for Soyabean Production and Yield Efficiency
plt.figure(figsize=(14, 7))

# Soyabean Production Bar Chart
plt.subplot(1, 2, 1)
ax1 = sns.barplot(
    x='SOYABEAN PRODUCTION (1000 tons)', 
    y='State Name', 
    data=top_5_soyabean, 
    palette='Blues_d'
)

plt.title('Top 5 States for Soyabean Production', fontsize=16, fontweight='bold')
plt.xlabel('Soyabean Production (1000 tons)', fontsize=12, fontweight='bold')
plt.ylabel('State Name', fontsize=12, fontweight='bold')

# Add data labels to bars
for p in ax1.patches:
    ax1.annotate(
        f'{p.get_width():.1f}', 
        (p.get_x() + p.get_width(), p.get_y() + p.get_height() / 2),
        ha='left', va='center', fontsize=12, fontweight='bold'
    )

# Yield Efficiency Bar Chart
plt.subplot(1, 2, 2)
ax2 = sns.barplot(
    x='Yield Efficiency (tons/hectare)', 
    y='State Name', 
    data=top_5_soyabean, 
    palette='Greens_d'
)

plt.title('Yield Efficiency of Top 5 Soyabean Producing States', fontsize=16, fontweight='bold')
plt.xlabel('Yield Efficiency (tons/hectare)', fontsize=12, fontweight='bold')
plt.ylabel('State Name', fontsize=12, fontweight='bold')

# Add data labels to bars
for p in ax2.patches:
    ax2.annotate(
        f'{p.get_width():.2f}', 
        (p.get_x() + p.get_width(), p.get_y() + p.get_height() / 2),
        ha='left', va='center', fontsize=12, fontweight='bold'
    )

plt.tight_layout()
plt.show()
```

```python
# Impact of Area Cultivated on Production (Rice, Wheat, Maize)
columns_of_interest = [
    'State Name', 
    'RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)', 
    'WHEAT AREA (1000 ha)', 'WHEAT PRODUCTION (1000 tons)', 
    'MAIZE AREA (1000 ha)', 'MAIZE PRODUCTION (1000 tons)'
]

crop_data = df[columns_of_interest]

plt.figure(figsize=(18, 6))

# Rice Production vs Area Cultivated
plt.subplot(1, 3, 1)
sns.scatterplot(
    x=crop_data['RICE AREA (1000 ha)'], 
    y=crop_data['RICE PRODUCTION (1000 tons)'], 
    color='blue'
)
plt.title('Rice Production vs Area Cultivated', fontsize=18, fontweight='bold')
plt.xlabel('Area Cultivated (1000 ha)', fontsize=14, fontweight='bold')
plt.ylabel('Rice Production (1000 tons)', fontsize=14, fontweight='bold')

# Wheat Production vs Area Cultivated
plt.subplot(1, 3, 2)
sns.scatterplot(
    x=crop_data['WHEAT AREA (1000 ha)'], 
    y=crop_data['WHEAT PRODUCTION (1000 tons)'], 
    color='green'
)
plt.title('Wheat Production vs Area Cultivated', fontsize=18, fontweight='bold')
plt.xlabel('Area Cultivated (1000 ha)', fontsize=14, fontweight='bold')
plt.ylabel('Wheat Production (1000 tons)', fontsize=14, fontweight='bold')

# Maize Production vs Area Cultivated
plt.subplot(1, 3, 3)
sns.scatterplot(
    x=crop_data['MAIZE AREA (1000 ha)'], 
    y=crop_data['MAIZE PRODUCTION (1000 tons)'], 
    color='orange'
)
plt.title('Maize Production vs Area Cultivated', fontsize=18, fontweight='bold')
plt.xlabel('Area Cultivated (1000 ha)', fontsize=14, fontweight='bold')
plt.ylabel('Maize Production (1000 tons)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Calculate correlation between Area and Production for each crop
rice_corr = crop_data['RICE AREA (1000 ha)'].corr(crop_data['RICE PRODUCTION (1000 tons)'])
wheat_corr = crop_data['WHEAT AREA (1000 ha)'].corr(crop_data['WHEAT PRODUCTION (1000 tons)'])
maize_corr = crop_data['MAIZE AREA (1000 ha)'].corr(crop_data['MAIZE PRODUCTION (1000 tons)'])

# Print the correlation values
print(f'Correlation between Rice Area and Production: {rice_corr:.2f}')
print(f'Correlation between Wheat Area and Production: {wheat_corr:.2f}')
print(f'Correlation between Maize Area and Production: {maize_corr:.2f}')
```

```python
# Year-wise Trend of Rice Production Across States (Top 3)

query = '''
        SELECT Year, State_Name, SUM(RICE_PRODUCTION) AS Total_Rice_Production
        FROM icrisat
        GROUP BY Year, State_Name
        ORDER BY Total_Rice_Production DESC
        LIMIT 3;
        
        '''

mycursor.execute(query)
results = mycursor.fetchall()

# Convert results into a DataFrame
answer_1 = pd.DataFrame(results, columns=["Year", "State_Name","Total_Rice_Production"])
print(answer_1)
```

```python
# Top 5 Districts by Wheat Yield Increase Over the Last 5 Years

query = '''

       SELECT Dist_Name, 
       (MAX(WHEAT_YIELD) - MIN(WHEAT_YIELD)) / MIN(WHEAT_YIELD) * 100 AS Growth_Rate
       FROM icrisat
       WHERE Year >= (SELECT MAX(Year) - 5 FROM icrisat)
       GROUP BY Dist_Name
       ORDER BY Growth_Rate DESC 
       LIMIT 5;
     
       '''

mycursor.execute(query)
results = mycursor.fetchall()

# Convert results into a DataFrame
answer_2 = pd.DataFrame(results, columns=["Dist_Name", "Growth_Rate"])
print(answer_2)
```

```python
# States with the Highest Growth in Oilseed Production (5-Year Growth Rate)

query = '''

       SELECT State_Name, 
       (MAX(OILSEEDS_PRODUCTION) - MIN(OILSEEDS_PRODUCTION)) / MIN(OILSEEDS_PRODUCTION) * 100 AS Growth_Rate
        FROM icrisat
       WHERE Year >= (SELECT MAX(Year) - 5 FROM icrisat)
       GROUP BY State_Name
       ORDER BY Growth_Rate DESC
       LIMIT 5;
     
       '''

mycursor.execute(query)
results = mycursor.fetchall()

# Convert results into a DataFrame
answer_3 = pd.DataFrame(results, columns=["State_Name", "Growth_Rate"])
print(answer_3)
```

```python
# District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)

query = '''
        SELECT Dist_Name,
       (SUM(RICE_AREA * RICE_PRODUCTION) - SUM(RICE_AREA) * SUM(RICE_PRODUCTION) / COUNT(*)) /
       (SQRT((SUM(RICE_AREA * RICE_AREA) - SUM(RICE_AREA) * SUM(RICE_AREA) / COUNT(*)) * 
             (SUM(RICE_PRODUCTION * RICE_PRODUCTION) - SUM(RICE_PRODUCTION) * SUM(RICE_PRODUCTION) / COUNT(*)))) AS Rice_Correlation,
       
       (SUM(WHEAT_AREA * WHEAT_PRODUCTION) - SUM(WHEAT_AREA) * SUM(WHEAT_PRODUCTION) / COUNT(*)) /
       (SQRT((SUM(WHEAT_AREA * WHEAT_AREA) - SUM(WHEAT_AREA) * SUM(WHEAT_AREA) / COUNT(*)) * 
             (SUM(WHEAT_PRODUCTION * WHEAT_PRODUCTION) - SUM(WHEAT_PRODUCTION) * SUM(WHEAT_PRODUCTION) / COUNT(*)))) AS Wheat_Correlation,
       
       (SUM(MAIZE_AREA * MAIZE_PRODUCTION) - SUM(MAIZE_AREA) * SUM(MAIZE_PRODUCTION) / COUNT(*)) /
       (SQRT((SUM(MAIZE_AREA * MAIZE_AREA) - SUM(MAIZE_AREA) * SUM(MAIZE_AREA) / COUNT(*)) * 
             (SUM(MAIZE_PRODUCTION * MAIZE_PRODUCTION) - SUM(MAIZE_PRODUCTION) * SUM(MAIZE_PRODUCTION) / COUNT(*)))) AS Maize_Correlation
       
       FROM icrisat
       GROUP BY Dist_Name;
       
       '''

mycursor.execute(query)
results = mycursor.fetchall()

# Convert results into a DataFrame
answer_4 = pd.DataFrame(results, columns=["Dist_Name", "Rice_Correlation", "Wheat_Correlation", "Maize_Correlation"])
print(answer_4)
```
